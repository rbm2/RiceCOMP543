In this exercise we'll be optimizing the [McCormick function](https://en.wikipedia.org/wiki/File:McCormick_function.pdf), one of a [set of common functions](https://en.wikipedia.org/wiki/Test_functions_for_optimization) used to benchmark optimization techniques.

The function $f : \mathbb{R}^2 \mapsto \mathbb{R}$ is defined as
$$
f(x, y) = \sin{(x + y)} + (x - y)^2 - 1.5x + 2.5y + 1
$$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Here is a python implementation of the function:

```python
def mccormick(point):
    x, y = point
    return np.sin(x + y) + (x - y)**2 - (1.5 * x) + (2.5 * y) + 1
```

And here is that function plotted over the domain $x, y$

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
ax = plt.figure().gca(projection="3d")
X = np.arange(-3, 3, 0.25)
Y = np.arange(-3, 3, 0.25)
X, Y = np.meshgrid(X, Y)
Z = mccormick((X, Y))
surf = ax.plot_surface(X, Y, Z, cmap="nipy_spectral")
print(f"Min Value: {np.min(Z)}")
```
