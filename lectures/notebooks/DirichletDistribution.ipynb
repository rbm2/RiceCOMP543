# The Dirichlet Distribution

Images from http://www.cs.cmu.edu/~epxing/Class/10701-08s/recitation/dirichlet.pdf

```python
import numpy as np
```

Dirichlet(1,1,1)

Values are uniformly sampled

```python
myD = np.random.dirichlet([1,1,1], 10)
myD
```

```python
np.var(myD)
```

Dirichlet(2,2,2)

```python
myD = np.random.dirichlet([2,2,2], 10)
myD
```

```python
np.var(myD)
```

Dirichlet(10,10,10)

Samples are more likely to be (1/3, 1/3, 1/3)

```python
myD = np.random.dirichlet([10,10,10], 10)
myD
```

```python
np.var(myD)
```

Samples are even more likely to be (1/3, 1/3, 1/3)

```python
myD = np.random.dirichlet([100,100,100], 10)
myD
```

```python
np.var(myD)
```

Dirichlet(2,10,2)

If a parameter is larger, that dimension get more weight. Notice the values in for the middle parameter.

```python
myD = np.random.dirichlet([2,10,2], 10)
myD
```

Dirichlet(0.9,0.9,0.9)

 When parameters are < 1, the samples go to the outside (more extreme differences). Note we have to round the values to get them out of scientific notation

```python
myD = np.random.dirichlet([.9, .9, .9], 10)
myD
```

```python
myD = np.around(np.random.dirichlet([.9, .9, .9], 10), 3)
myD
```

```python
np.var(myD)
```
